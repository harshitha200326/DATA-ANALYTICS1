```python
# Set style for Seaborn
sns.set(style='whitegrid')
 # Other options: 'darkgrid', 'white', 'dark'

# Set Matplotlib defaults
plt.rcParams['figure.figsize'] = (12, 6)
# Width, height
plt.rcParams['axes.titlesize'] = 16
# Title font size
plt.rcParams['axes.labelsize'] = 14
 # Axis label font size
plt.rcParams['xtick.labelsize'] = 12
# X-axis tick label font
plt.rcParams['ytick.labelsize'] = 12
# Y-axis tick label font
plt.rcParams['legend.fontsize'] = 12
# Legend font size
plt.rcParams['figure.dpi'] = 100
# Resolution
```

```python
sns.countplot(data=df, y='Status', order=df['Status'].value_counts().index)
plt.title("Order Status Distribution")
plt.xlabel("Number of Orders")
plt.ylabel("Order Status")
plt.tight_layout()
plt.show()
```

```python
# Select only numerical columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# Create a pivot table
pivot = pd.pivot_table(df, values='Qty', index='Category', columns='ship-state', aggfunc='sum', fill_value=0)

# Plot the heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(pivot, annot=False, cmap='Blues', linewidths=0.3)
plt.title("Heatmap of Quantity Sold by Category and State")
plt.xlabel("Ship State")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()
```

```python
# Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y', errors='coerce')

# Fill missing values
df['Amount'] = df['Amount'].fillna(0)
df['Qty'] = df['Qty'].fillna(0)

# Set default styles
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# 1. Order Status Distribution
plt.figure()
sns.countplot(data=df, y='Status', order=df['Status'].value_counts().index)
plt.title("Order Status Distribution")
plt.xlabel("Number of Orders")
plt.ylabel("Status")
plt.tight_layout()
plt.show()
```

```python
##Top 5 Categories by Sales
top_categories = df.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(5)
plt.figure()
top_categories.plot(kind='bar', color='skyblue')
plt.title("Top 5 Categories by Total Sales")
plt.ylabel("Total Sales (INR)")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Top 5 Cities by Order Count
top_cities = df['ship-city'].value_counts().head(5)
plt.figure()
top_cities.plot(kind='bar', color='orange')
plt.title("Top 5 Cities by Order Count")
plt.ylabel("Number of Orders")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print(df['Date'].nunique())
```

```python
#Monthly Sales Trend
df['Month'] = df['Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Amount'].sum()
plt.figure()
monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Total Sales (INR)")
plt.xlabel("Month")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
##Cancelled Orders per Category
cancelled = df[df['Status'].str.lower().str.contains('cancelled')]
cancelled_per_category = cancelled['Category'].value_counts()
plt.figure()
cancelled_per_category.plot(kind='bar', color='red')
plt.title("Cancelled Orders per Category")
plt.ylabel("Number of Cancellations")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
